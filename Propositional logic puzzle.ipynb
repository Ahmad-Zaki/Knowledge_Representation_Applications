### propositional statements:

 - $R$ : Red door is safe      
 - $G$ : Green door is safe    
 - $B$ : Blue door is safe     
 - $S1$: Only one door is safe = $(RG'B')$ $OR$ $(R'GB')$ $OR$ $(R'G'B)$  
***
 - $X$ : Red door is safe = $R$         
 - $Y$ : Blue door is not safe = $B'$    
 - $Z$ : Blue is not safe = $B'$    
 - $S2$:At least one is true & at least one is false = $(XY'Z')$ $OR$ $(X'YZ')$ $OR$ $(X'Y'Z)$ $OR$ $(XYZ')$ $OR$ $(XY'Z)$ $OR$ $(X'YZ)$  

```python
#Generate all possible Combinations of R,G,B: 
from itertools import product

com = list(product([0,1],[0,1],[0,1]))
print(com)
```

```python
#Evaluate S1:
s1= lambda R,G,B : (R&(not G)&(not B))|((not R)&G&(not B))|((not R)&(not G)&B) 

#Evaluate S2:
s2 = lambda X,Y,Z: (X&(not Y)&(not Z))|((not X)&Y&(not Z))|((not X)&(not Y)&Z)|(X&Y&(not Z))|(X&(not Y)& Z)|((not X)&Y&Z)
```

```python
def solution(com):
    for i in com:
        X,Y,Z = i[0], not i[2], not i[2]
#         print(f"{i}: {s1(*i)}, {s2(X,Y,Z)}")
        if s1(*i) and s2(X,Y,Z):
            print(f'the {dict([((1,0,0),"red"),((0,1,0),"green"),((0,0,1),"blue")])[i]} door will lead to safety')
            
```
